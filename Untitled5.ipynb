```python
df['Gender'].fillna(df['Gender'].mode()[0])
df['Smoking_Status'].fillna(df['Smoking_Status'].mode()[0])
df['BMI'].fillna(df['BMI'].mean())
df['Blood_Pressure'].fillna(df['Blood_Pressure'].mean())
df['Cholesterol'].fillna(df['Cholesterol'].mean())


df.isnull().sum()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Print how many outliers we remove
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    print(f"{column}: Removed {outliers.shape[0]} outliers")
    # Remove them
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Apply outlier removal for each numerical column
for col in continious_var:
    df = remove_outliers_iqr(df, col)


df.head()
```
