```python
import pandas as pd 
df=pd.read_csv("area_price_dataset.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt  
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
x=df[['area']]
y=df['price']
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train) #actually produces the linear eqn for the data
pred_val=regressor.predict(X_test)
```

```python
print("===============================")
print("coeff is")
print(regressor.coef_)
print("intercept is")
print(regressor.intercept_)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(X_train, regressor.predict(X_train), color='blue') # plotting the regression line
plt.show()
```

```python
# Calculate Mean Squared Error and R-squared
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

mse = mean_squared_error(y_test, pred_val)
mae = mean_absolute_error(y_test, pred_val)
r2 = r2_score(y_test, pred_val)
print("mean_squared_error",mse)
print("mean_absolute_error",mae)
print("R2",r2)
print("R2",regressor.score(x,y))
```

```python
# Adjusted R² calculation
n = x.shape[0]  # number of samples
p = x.shape[1]  # number of features
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("R²:", r2)
print("Adjusted R²:", adjusted_r2)
```
